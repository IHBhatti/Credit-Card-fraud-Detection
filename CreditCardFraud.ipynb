The given dataset is unbalanced in which we analysed that if we train the model on that data it will show approximately every transaction as a normal because the given dataset contains huge difference between normal and fraudulent transactions. Where 0 shows normal transaction and 1 shows fraudulent transaction.

```python
normal=ds[ds.Class==0]
fraud=ds[ds.Class==1]
print(normal.shape)
print(fraud.shape)
```

**Splitting the dataset **

```python
x=new_ds.drop('Class', axis=1)
y=new_ds['Class'] # This is the column that is intend to be predicted.
# Use x and y variables to split the Dataset into train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=.20, stratify=y, random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

```python
# Importing Model
from sklearn.linear_model import LogisticRegression
# Create model
M1=LogisticRegression()
# Passing Training data to model
M1.fit(x_train, y_train)
```
